```python
import pandas as pd
import os
```

Task 1: Merging 12 months of data in single csv file

```python
df=pd.read_csv("./Sales_Data/Sales_April_2019.csv")

df.head()
```

```python
files=[file for file in os.listdir('./Sales_Data')]

for file in files:
    print(file)
```

```python
all_months_data=pd.DataFrame()
for file in files:
    df=pd.read_csv("./Sales_Data/"+file)
    all_months_data=pd.concat([all_months_data,df])
    
all_months_data.to_csv("all_data.csv", index=False)
```

Read updated data frame

```python
all_data=pd.read_csv('all_data.csv')
all_data.head()
```

### Cleaning of Data

##### Drop rows of NAN 

```python
nan_df=all_data[all_data.isna().any(axis=1)]
nan_df.head()
```

```python
all_data=all_data.dropna(how='all')
all_data.head()
```

###  Find O and delete

```python
temp_df=all_data[all_data['Order Date'].str[0:2]=='Or']
temp_df.head()
```

```python
all_data=all_data[all_data['Order Date'].str[0:2]!='4/']
all_data.head()
```

```python
all_data=all_data[all_data['Order Date'].str[0:2]!='5/']
all_data.head()
```

```python
all_data=all_data[all_data['Order Date'].str[0:2]!='Or']
all_data.head()
```

#### Convert Column to correct type

```python
all_data['Quantity Ordered']=pd.to_numeric(all_data['Quantity Ordered']) #make int
all_data['Price Each']=pd.to_numeric(all_data['Price Each']) #make float
all_data.head()
```

# Augment Data with Additional Columns

### Task 2: Adding Month column:

```python
all_data['Month']=all_data['Order Date'].str[0:2]
all_data['Month']=all_data['Month'].astype('int32')

all_data.tail()
```

### Task 3: ADD SALES COLUMN

```python
all_data['Sales']=all_data['Quantity Ordered']*all_data['Price Each']
all_data.head()
```

### TASK4: ADD A CITY COLUMN

```python
#let us use .apply() and lambda
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]
all_data['City']=all_data['Purchase Address'].apply(lambda x: get_city(x)+ ' (' + get_state(x) +')')
all_data.head()
```

# Question1:-What was the best month for sales?How much was earned that month?

```python
results=all_data.groupby('Month').sum()
results
```

```python
import matplotlib.pyplot as plt
months=range(1,13)
plt.bar(months,results['Sales'])
plt.xticks(months)
plt.xlabel('Months Number')
plt.ylabel('Sales in USD ($)')
plt.show()
```

#### QUESTION 2: WHICH CITY HAS HIGER NUMBER OF SALES

```python
result=all_data.groupby('City').sum()
result
```

```python
import matplotlib.pyplot as plt

cities=[city for city,df in all_data.groupby('City')]
plt.bar(cities,result['Sales'])
plt.xticks(cities, rotation='vertical',size=8)
plt.xlabel('City Name')
plt.ylabel('Sales in USD ($)')
plt.show()
```

### QUESTION3:WHAT TIME SHOULD WE DISPLAY ADS TO INCREASE THE SALES?

```python
all_data.head()
```

```python
all_data['Order Date']=pd.to_datetime(all_data['Order Date'])
all_data.head()
```

```python
all_data['Hour']=all_data['Order Date'].dt.hour
all_data['Minute']=all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours=[hour for hour,df in all_data.groupby('Hour')]
plt.plot(hours,all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hours')
plt.ylabel('Numbers Of Orders')
plt.grid()
plt.show()

#my recomendation is to display ads around 11 at morning and 7 at evening
```

### QUE4: WHAT PRODUCTS ARE MOST OFTEN SOLD TOGETHER?

```python
df=all_data[all_data['Order ID'].duplicated(keep=False)]
df.head(20)
```

```python
df['Grouped']=df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head()
```

```python
df=df[['Order ID' , 'Grouped']].drop_duplicates()
df.head(50)
```

```python
from itertools import combinations
from collections import Counter
```

```python
count=Counter()
for row in df['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
count.most_common(5)
```
